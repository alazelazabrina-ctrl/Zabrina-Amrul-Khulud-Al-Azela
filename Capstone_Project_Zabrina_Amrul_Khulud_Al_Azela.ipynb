```python
df = pd.read_excel("Tingkat Pengangguran Terbuka Menurut Provinsi, 2025.xlsx", skiprows=3)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Remove the last few rows which seem to be summary or footnotes
df_cleaned = df.iloc[:38].copy()

# Rename columns for clarity
df_cleaned = df_cleaned.rename(columns={'Unnamed: 0': 'Provinsi', 'Februari': 'Februari_2025', 'Agustus': 'Agustus_2025', 'Tahunan': 'Tahunan_2025'})

# Convert relevant columns to numeric, coercing errors to NaN
df_cleaned['Februari_2025'] = pd.to_numeric(df_cleaned['Februari_2025'], errors='coerce')
df_cleaned['Agustus_2025'] = pd.to_numeric(df_cleaned['Agustus_2025'], errors='coerce')
df_cleaned['Tahunan_2025'] = pd.to_numeric(df_cleaned['Tahunan_2025'], errors='coerce')

# Display the cleaned dataframe and its info
display(df_cleaned.head())
display(df_cleaned.info())
```

# Task
Perform data cleaning on the dataframe `df` by handling string and numeric data types appropriately, adjusting column names for consistency, and then sort the data by the `Februari_2025` column in descending order. Finally, create a bar plot to visualize the sorted data. Use the file "Tingkat Pengangguran Terbuka Menurut Provinsi, 2025.xlsx".

## Sort data

### Subtask:
Sort the cleaned DataFrame `df_cleaned` by the 'Februari_2025' column in descending order.


**Reasoning**:
Sort the dataframe by 'Februari_2025' in descending order and display the head.



```python
df_cleaned = df_cleaned.sort_values(by='Februari_2025', ascending=False)
display(df_cleaned.head())
```

## Create bar plot

### Subtask:
Create a bar plot using the sorted data to visualize the unemployment rate per province.


**Reasoning**:
Create a bar plot to visualize the sorted unemployment rate per province using the cleaned dataframe.



```python
plt.figure(figsize=(15, 8))
plt.bar(df_cleaned['Provinsi'], df_cleaned['Februari_2025'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('Provinsi')
plt.ylabel('Tingkat Pengangguran Terbuka Februari 2025 (%)')
plt.title('Tingkat Pengangguran Terbuka per Provinsi Februari 2025 (Sorted)')
plt.tight_layout()
plt.show()
```

## Add labels and title

### Subtask:
Add labels to the x and y axes and a title to the plot for clarity.


**Reasoning**:
Add labels to the x and y axes and a title to the plot for clarity.



```python
plt.figure(figsize=(15, 8))
plt.bar(df_cleaned['Provinsi'], df_cleaned['Februari_2025'], color='skyblue')
plt.xticks(rotation=90)
plt.xlabel('Provinsi')
plt.ylabel('Tingkat Pengangguran Terbuka Februari 2025 (%)')
plt.title('Tingkat Pengangguran Terbuka per Provinsi Februari 2025')
plt.tight_layout()
plt.show()
```
